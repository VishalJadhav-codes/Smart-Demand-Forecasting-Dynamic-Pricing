# Create ML-ready features

```python
import pandas as pd

df = pd.read_csv(r'C:\Projects\Smart Demand Forecasting & Dynamic Pricing Recommendation System\data\smart_demand_pricing_dataset.csv')

# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])

# Create time-based features
df['day'] = df['date'].dt.day
df['month'] = df['date'].dt.month
df['dayofweek'] = df['date'].dt.dayofweek

# Sort values
df = df.sort_values("date")

# Lag features
#df['lag_1'] = df['units_sold'].shift(1)
df['lag_7'] = df['units_sold'].shift(7)

# Rolling features
#df['rolling_7'] = df['units_sold'].rolling(7).mean()
df['rolling_14'] = df['units_sold'].rolling(14).mean()

# Drop NA
df.dropna(inplace=True)

# Save features data
df.to_csv(r"C:\Projects\Smart Demand Forecasting & Dynamic Pricing Recommendation System\data\feature_data.csv", index=False)
```
